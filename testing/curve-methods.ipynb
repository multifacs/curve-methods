```python
# imports
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
```

```python
class Section:
    def __init__(self):
        self.a = Point(0, 0)
        self.b = Point(0, 0)

    def get_x(self, t):
        return self.a.x + t*(self.b.x - self.a.x)

    def get_y(self, t):
        return self.a.y + t*(self.b.y - self.a.y)
```

```python
s = Section()
s.a = Point(1, 1)
s.b = Point(3, 5)

x = [s.get_x(t) for t in np.arange(0, 1, 0.05)]
y = [s.get_y(t) for t in np.arange(0, 1, 0.05)]
```

```python
plt.figure(figsize=(6, 4))
plt.scatter(x, y, marker=".")
```

```python
class Ellipsis:
    def __init__(self):
        self.c = Point(0, 0)
        self.a = 0
        self.b = 0
        self.phi = 0

    def get_x(self, t):
        return self.c.x + self.a * np.cos(t) * np.cos(self.phi) - self.b * np.sin(t) * np.sin(self.phi)

    def get_y(self, t):
        return self.c.y + self.a * np.cos(t) * np.sin(self.phi) + self.b * np.sin(t) * np.cos(self.phi)
```

```python
e = Ellipsis()
e.c = Point(1, 1)
e.a = 3
e.b = 5
e.phi = np.pi * 0.2

x = [e.get_x(t) for t in np.arange(0, np.pi * 2, 0.05)]
y = [e.get_y(t) for t in np.arange(0, np.pi * 2, 0.05)]
```

```python
plt.figure(figsize=(6, 4))
plt.scatter(x, y, marker=".")
```

```python
class SineWave:
    def __init__(self):
        self.a = 0
        self.t = 0
        self.c = 0

    def get_x(self, t):
        return t

    def get_y(self, t):
        return self.a + np.sin(np.pi * 2 / self.t * t) + self.c
```

```python
si = SineWave()
si.c = 1
si.a = 3
si.t = 5

x = [si.get_x(t) for t in np.arange(0, np.pi * 2, 0.05)]
y = [si.get_y(t) for t in np.arange(0, np.pi * 2, 0.05)]
```

```python
plt.figure(figsize=(6, 4))
plt.scatter(x, y, marker=".")
```
